## Loading Required Libraries

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

### Loading Dataset

```python
fifa = pd.read_csv("players_20.csv")
fifa.head()
```

##### Print all columns in our dataset

```python
for col in fifa.columns:
    print(col)
```

#### Query the number of rows and columns in the dataset

```python
fifa.shape
```

##### Get count of a specific column eg nationality

```python
fifa['nationality'].value_counts()
```

#### Query the first 10 records of top countries with most players

```python
fifa['nationality'].value_counts()[0:10]
```

#### Query the first 5 records of top countries with most players

```python
fifa['nationality'].value_counts()[0:5]
```

```python
fifa['nationality'].value_counts()[0:5].keys() # index can stand for keys().....[0:5].index
```

### Bar Plot for top 5 countries with most players

```python
plt.figure(figsize=(8,5))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()),list(fifa['nationality'].value_counts()[0:5]),color='g')
plt.show()
```

### Query players' salary based on their short name column and wage_eur

```python
player_salary = fifa[['short_name','wage_eur']]
player_salary.head()
```

### Find the highest played players from the above list

```python
player_salary = player_salary.sort_values(by=['wage_eur'],ascending=False)
player_salary.head()
```

### Make a bar plot to show distribution of categorical entities

```python
plt.figure(figsize=(8,5))
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'])[0:5],color=['pink','orange','blue','red','cyan'])
plt.show()
```
